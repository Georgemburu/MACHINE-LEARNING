```python
# vectorize
class Vect_tickets:
    def __init__(self,dframecolumn):
        self.indices_ticket = dict()
        self.ticket_indices = dict()
        counter = 0;
        for tkt in dframecolumn:
            if counter == 0:
                print('tkt',tkt)
            if tkt not in self.ticket_indices:
                counter = counter + 1
                self.ticket_indices[tkt] = counter;
        
        for k in self.ticket_indices:
            valuee = self.ticket_indices[k]
            self.indices_ticket[valuee] = k
        
#         print('indices_ticket',self.indices_ticket)
#         print('self.ticket_indices',self.ticket_indices)
    
    def encode(self,dframecol):
        correctedDfCol = []
        for c in iter(dframecol):
            correctedDfCol.append(self.ticket_indices[c])
        
        return np.array(correctedDfCol)
    
    def decode(self,dframecoltoDecode):
        correctedDfCol2 = []
        for c in iter(dframecol):
            correctedDfCol2.append(self.indices_ticket[c])
        
        return correctedDfCol2
            
    
        
        
        
        
        
        
```

```python
not_encoded_tickets_DF = df['ticket']

vect_tickets = Vect_tickets(not_encoded_tickets_DF)
encoded_tickets = vect_tickets.encode(not_encoded_tickets_DF)
print(len(not_encoded_tickets_DF),len(encoded_tickets))
df['ticket'] = encoded_tickets
df
```
