```python
from sklearn import cluster, datasets
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target

k_means = cluster.KMeans(n_clusters=3)
fit = k_means.fit(X_iris)
print(fit)
print('pred labels',k_means.labels_[::10])
print('true labels', y_iris[::10])
```

```python
# import matplotlib.pyplot as plt
plt.plot(X_iris,'o')
plt.plot(k_means.cluster_centers_,'-')
plt.plot(k_means.labels_,c='g')

a = X_iris.shape
# import
b = k_means.labels_.shape
print(a,b)
# dir(k_means)
```
