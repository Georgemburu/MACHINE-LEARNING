```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
import pandas as pd
```

```python
# IRIS DATASET
#=> contains 4 features and 1 label
#> features: -> sepal length
#            -> sepal width
#            -> petal length
#            -> petal width
```

```python
CSV_COLUMN_NAMES = ['SepalLength','SepalWidth','PetalLength',
                   'PetalWidth','Species']
SPECIES = ['Setosa','Versicolor','Virginica']
```

```python
# Download and parse the data
train_path = tf.keras.utils.get_file(
    'iris_training.csv','https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv'
)
test_path = tf.keras.utils.get_file(
    'iris_test.csv','https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv'
)
train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)
```

```python
train.head()
```

```python
# For each of the datasets, split out the labels, 
# which the model will be trained to predict
train_y = train.pop('Species')
test_y = test.pop('Species')

# The label column has now been removed from the features
train.head()
```

```python
#  Overview of programming with estimators
# An estimator is: Any class derived from (tf.estimator.Estimator)
# To write a Tf program based on pre_made Estimators, you 
# must perform the following tasks.
# 1. Create one or more input functions
# 2. Define the model's feature columns
# 3. Instantiate an Estimator, specifying the feature columns
# and various hyperparameters.
# 4. Call one or more methods on the Estimator object, passing
# the  appropriate input function as the source of the data
```

```python
# Create input functions
#############################
# An input function is a function that returns (tf.data.Dataset)
# object which outputs the following two_element turple:
# A. (features) - A Python dict in which
#  - Each key is the name of a feature
#  - Each value is an array containing all of that feature's 
#  values
# B. (label) - An array containing the values of the label
#  for every example
##################
# Format of an input function

def input_evaluation_set():
    features = {
        'SepalLength': np.array([6.4,5.0]),
        'SepalWidth': np.array([2.8, 2.3]),
        'PetalLength': np.array([5.6, 3.3]),
        'PetalWidth': np.array([2.2, 1.0])
    }
    labels = np.array([2,1])
    return features, labels
```

```python
def input_fn(features, labels, training=True, batch_size=256):
    '''An input function for training or evaluating'''
    # Convert the inputs to a Dataset
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))
    print('converted dataset=', dataset)
    # shuffle and repeat if you are in training mode
    if(training):
        dataset = dataset.shuffle(1000).repeat()
    
    return dataset.batch(batch_size)
```

```python
# Define the feature columns
####################################
# A feature column is an object describing how the model should
# use raw input data from the feature dictionary
# The (tf.feature_column) module provides many options
# for representing data to the model
# For Iris, the 4 raw features are numeric values, so we'll 
# build a list of feature columns to tell the Estimator model
# to represent each of the four features as 32-bit floating-
# point value

# Feature columns describe how to use the input.
my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))

my_feature_columns
```

```python
# Instatiate an Estimator
#################################
#  The iris problem is a classification problem
# Tf Pre-made classifier Estimators
# (tf.estimator.DNNClassifier) -  for deep models that 
# perform mult-class classificaton
# - (tf.estimator.DNNLinearCombinedClasssifier) - for wide & 
# deep models
# - (tf.estimator.LinearClassifier) - for classifiers based
# on linear models

# We'll use (tf.estimator.DNNCLassifier)

# BUILD a DNN with 2 hidden layers with 30 and 10 hidden nodes
# each
classifier =  tf.estimator.DNNClassifier(
    feature_columns=my_feature_columns,
    # Two hidden layers of 30 and 10 nodes respectively
    hidden_units = [30,10],
    # The model must choose between 3 classes.
    n_classes=3
)
```

```python
# Train, Evaluate  and Predict
###############
# Train the model
classifier.train(
    input_fn=lambda: input_fn(train, train_y, training=True),
    steps=5000
)
```

```python
# Evaluate the trained model
eval_results = classifier.evaluate(
    input_fn = lambda: input_fn(test,test_y, training=False)
)
print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**eval_results))
```

```python
# Making predictions (inferring) from the trained model
# Generate predictions from the model
expected = ['Setosa','Versicolor','Virginica']
predict_x = {
    'SepalLength': [5.1, 5.9,6.9],
    'SepalWidth': [3.3,3.0,3.1],
    'PetalLength':[1.7,4.2,5.4],
    'PetalWidth':[0.5,1.5,2.1]
}

def input_fn(features, batch_size=256):
    '''An input function for prediction.'''
    # Convert the inputs to a Dataset without labels.
    return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

predictions = classifier.predict(
    input_fn=lambda: input_fn(predict_x)
)
```

```python
# The (predict) method returns a Python iterable, yielding a 
# dict of prediction results for each example
for pred_dict, expec in zip(predictions, expected):
    class_id = pred_dict['class_ids'][0]
    probability = pred_dict['probabilities'][class_id]
    
    print('Prediction is "{}" ({:.1f}%), expected "{}"'.format(
        SPECIES[class_id], 100 * probability, expec
    ))
```
