```python
from keras.models import Sequential,load_model
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Dropout
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = image.load_img('img.jpeg')
img_arr = image.img_to_array(img)
img_arr.shape
img_arr = np.expand_dims(img_arr,axis=0)
img_arr.shape
# np.
```

```python
plt.imshow(img)
```

```python
model = Sequential()
model.add(Conv2D(4,(6,6),input_shape=(330,330,3)))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(4,(2,2)))
model.add(MaxPooling2D(pool_size = (1,1)))
model.add(Dropout(rate=0.5))
# model.add(Flatten())
# model.add(Dense(32, input_shape=(330,300,3)))
model.compile('adam',loss='sparse_categorical_crossentropy')
model.summary()
```

```python
# model.save('resize_to_(161*161)_and_grayscale.h5')
```

```python
pred = model.predict(img_arr)
# plt.imshow(pred[0][0:-1])
print(pred[0][0:].shape)
# plt.imshow(image.array_to_img(pred[0][0:]))
image.array_to_img(pred[0][0:])
```
